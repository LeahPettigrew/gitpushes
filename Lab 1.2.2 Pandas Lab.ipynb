#### 5.A Plot Histogram of Age 

````python
# A:
num_bins = 50
n, bins, patches = plt.hist(df.age, num_bins, facecolor='aqua', alpha =0.5)
````

#### 5.B BoxPlot of Age

````python
# A:
plt.boxplot(df.age)
plt.yscale('linear')
plt.grid(b=True, which='major', color='g', linestyle='-')
plt.grid(b=True, which='minor', color='y', linestyle='--')
plt.show()
````
